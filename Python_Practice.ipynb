**Read Data**

```python
import pandas as pd
import numpy as np
```

```python
car_info = pd.read_csv('/content/yt_car_info.csv')
car_main = pd.read_csv('/content/yt_car_main.csv')
car_used_price = pd.read_csv('/content/yt_car_used_prices.csv')
```

**What is the average retail price for Make and Model for the Years 2023 and 2024, premium Trim and horsepower > 400**

```python
#Fetch Index of Horsepower >400
car_info[['hp','rpm']]=car_info['Horsepower'].str.extract(r'(\d+)\s*hp\s*@\s*(\d+)\s*rpm').fillna(0).astype(int)
index_list = car_info[car_info['hp']>400]['index']

#Yeasr 2023, 2024
car_main[(car_main['Year'].isin([2023,2024])) & (car_main['Trim'].str.contains('Premium',case=False,na=False))  & (car_main['index'].isin(index_list))].groupby(['Make','Model'])['Retail_Price'].mean().reset_index()
```

**Create a Rank column based on the Invoice Price per Car in the main table - Order DESC**

```python
car_main['Invoice Price'] = car_main['Invoice Price'].str.replace('$','').str.replace(',','').astype(float)
```

```python
car_main['Car Rank'] = car_main['Retail_Price'].rank(method='dense',ascending=False)
car_main.sort_values(by='Car Rank')
```

**Show the total retail Price, average used price, cylinders and Horsepower per model that has Highway FE <34mpg**

```python
car_info['FE'] = car_info['Highway Fuel Economy'].str.extract('(\d+)\s*mpg\s*')
car_info['FE'] = car_info['FE'].fillna(-1).astype('int')
index_list = car_info[(car_info['FE']<34) & (car_info['FE']>0)]['index']
```

```python
car_main = pd.merge(car_main,car_used_price,on='index',how='left')
```

```python
car_main = pd.merge(car_main,car_info,on='index',how='left')
```

```python
results = car_main.groupby(['Model','hp','Cylinders'])[['Retail_Price','Used Price']].agg({'Retail_Price':sum,'Used Price':np.mean}).reset_index().rename({'Retail_Price':'Total Retail_Price','Used Price':'Avg Used Price'},axis=1)
```

**Create a new Category for the retail price using the following <50k, 50k-100k, >100k. Show the number of cars per Make and Category. Bonus for a Pivot table output**

```python
car_main['Category']=np.where(car_main['Retail_Price']<50000,'<50k',np.where(car_main['Retail_Price']<=100000,'50k-100k','>100k'))
```

```python
car_main.head(3)
```

```python
final_data = car_main.groupby(['Make','Category'])['index'].count().reset_index()
final_data.pivot(values='index',index='Make',columns='Category')
```
